```python
from sklearn.datasets import load_iris
```

```python
X_iris , Y_iris = load_iris(return_X_y=True)
```

```python
X_iris
```

```python
Y_iris
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X_iris,Y_iris,test_size=0.2, random_state=50)
```

```python
print("%d"%len(X_train))
print("%d"%len(Y_train))
print("%d"%len(X_test))
print("%d"%len(Y_test))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model_tree=DecisionTreeClassifier()
model_tree.get_params()
```

```python
model_tree.fit(X_train,Y_train)
```

```python
print("do chinh xac cua tap huan luyen",round(model_tree.score(X_train,Y_train)*100,2))
```

```python
feature_names=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
target_names = ['0: Setosa', '1: Versicolor', '2:Virginica']
```

```python
from sklearn import tree
text_tree= tree.export_text(model_tree,
                      feature_names=feature_names,
                      show_weights=True)
print(text_tree)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(25,20))
tree.plot_tree(model_tree,
              feature_names=feature_names,
              class_names=target_names,
              filled=True)
plt.show()
```

```python
print(model_tree.feature_importances_)
#trutc quan hoa
plt.figure(figsize=(8,4))
plt.barh(feature_names,model_tree.feature_importances_)
plt.title("muc do quan trong cua thuoc tinh")
plt.ylabel("danh sach thuoc tinh")
plt.xlabel("do quan trong")
plt.grid=True
plt.show()
```

```python
print('do sau cay quyet dinh',model_tree.get_depth())
```

```python
print('so nut la cua cay quyet dinh',model_tree.get_n_leaves())
```

```python
y_pred=model_tree.predict(X_test)
print(y_pred)
print(Y_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc=accuracy_score(Y_test,y_pred)
acc_num=accuracy_score(Y_test,y_pred,normalize=False)
print(round(acc*100,2))
print(acc_num,"/",len(Y_test))
```

```python
#thay doi thong so mo hinh
model_tree1=DecisionTreeClassifier(criterion='entropy',
                                  splitter='random',
                                  max_features=3,
                                  random_state=1)
```

```python
model_tree1.fit(X_train,Y_train)
print(round(model_tree.score(X_train,Y_train)*100,2))
```

```python
plt.figure(figsize=(8,4))
plt.barh(feature_names,model_tree1.feature_importances_)

plt.title('MỨC ĐỘ QUAN TRỌNG CỦA CÁC THỘC TÍNH')
plt.ylabel('Danh sách thuộc tính')
plt.xlabel('Độ quan trọng')
#plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(25,20))

tree.plot_tree(model_tree1, 
                   feature_names=feature_names,  
                   class_names=target_names,
                   filled=True)
plt.show()
```

```python
#Xác định độ sâu của cây: Khoảng cách tối đa giữa gốc và lá
print('Độ sâu của cây quyết định:',
      model_tree1.get_depth())

#Xác định số lượng nút lá của cây quyết định:
print('Số nút lá của cây quyết định:',
      model_tree1.get_n_leaves())
```

```python
y_pred1=model_tree.predict(X_test)
print("Kết quả phân loại hoa lan trên tập dữ liệu TEST:")
print("Predicted labels: ", y_pred1)
print("Ground truth    : ", Y_test)
```

```python
acc1=accuracy_score(Y_test,y_pred1)
acc1_num=accuracy_score(Y_test,y_pred1,normalize=False)
print("Độ chính xác đạt:", round(acc1*100,2))
print("Tổng số mẫu dự đoán đúng:",acc1_num, '/', len(Y_test))
```

```python
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(model_tree1,X_test,Y_test)
plt.show()
```
